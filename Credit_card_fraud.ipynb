```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
```

```python
models=[]
results=[]
names=[]
```

```python
models.append(('LR',LogisticRegression()))
models.append(('LDA',LinearDiscriminantAnalysis()))
models.append(('KNN',KNeighborsClassifier()))
models.append(('CART',DecisionTreeClassifier()))
models.append(('NB',GaussianNB()))
models.append(('SVM',SVC()))

```

```python
models
```

```python
from sklearn.impute import SimpleImputer
from sklearn import model_selection
import numpy as np

# Assuming `models`, `x`, and `y` are already defined
# Example definition:
# models = [('Logistic Regression', LogisticRegression()), ('SVM', SVC())]
# x, y = some_data_loading_function()

# Handle missing values by imputing them
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
x_imputed = imputer.fit_transform(x)

results = []
names = []

for name, model in models:
    kfold = model_selection.KFold(n_splits=10, random_state=7, shuffle=True)
    # Set error_score='raise' to get more informative error messages
    cv_results = model_selection.cross_val_score(model, x_imputed, y, cv=kfold, scoring='accuracy', error_score='raise')
    results.append(cv_results)
    names.append(name)
    msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
    print(msg)
```
